```python
import pandas as pd
import numpy as np
```

```python
amazonDataFrame = pd.read_csv('./data/Alexa-Top-1M.csv',names=['index','Domain'])
amazonDataFrame.drop(['index'],axis=1,inplace=True) #drop index
amazonDataFrame.head()
```

```python
from pronounceable import Pronounceablity
import nltk
nltk.download('cmudict')
```

```python
## Pronounceablity MIT License Algorithm
## https://github.com/patarapolw/pronounceable

pr = Pronounceablity()
pr.syllable('.')
```

```python
test_df = amazonDataFrame.iloc[0:100000]
```

```python
test_df['FLD'] = test_df['Domain'].str.split('.').str[-1]
test_df['SLD'] = test_df['Domain'].str.split('.').str[-2]
test_df['TLD'] = test_df['Domain'].str.split('.').str[-3]
```

```python
test_df.head()
```

```python
# test_df['len_SLD'] = test_df['SLD'].str.len()
test_df[(test_df['len_SLD'] == 2) & (test_df['SLD'] == 'co')].head()
```

```python
## .com 만 추출
test_com = test_df[test_df['FLD'] == 'com']
```

```python
test_com.reset_index(inplace=True)
```

```python
test_com.drop(['index'],axis=1)
```

```python
test_com['pronounce'] = 0
```

```python
point= pr.syllable(test_com['SLD'][0])
test_com['pronounce'][0]
```

```python
test_com['SLD'][11]
```

```python
for i in range(len(test_com['pronounce'])) :
    print(i)
    point= pr.syllable(test_com['SLD'][i])
    test_com['pronounce'][i] = point
    if (i % 10000 == 0) : print(i)
```

```python
for i in range(len(test_df['Domain'])) :
    temptext = test_df['Domain'][i]
    if temptext.split('.')[-1] == 'com' : 
        temptext = temptext.split('.')[-2]
        test_df['SLD'][i] = temptext


# ('aero' | 'arpa' | 'asia' | 'biz' | 'cat' | 'com' | 'coop' | 'edu' | 'gov' | 'info' | 'int' | 'jobs' |'mil' | 'mobi' | 'museum' | 'name' | 'net' | 'org' | 'pro' | 'tel' | 'travel' | 'wiki' | 'xxx' | 'xyz') 

# aero 
# arpa 도메인 시스템의 기술적 하부 구조 (=Address and Routing Parameter Area)
# asia 아시아 지역을 대표
# biz 사업
# cat 카탈로니아 언어/문화
# com 영리 목적의 기업이나 단체, .co.국가로 쓰기도 한다. ex)www.url.co.kr,www.url.co.jp
# coop 조합
# edu 미국의 4년제 이상 교육기관.
# gov 미국의 연방 정부 조직, .go.국가로 쓰기도 한다. ex)www.url.go.kr
# info 정보 관련 (무제한)
# int 국제 조약 등으로 만들어진 국제 기관
# jobs 취업 관련 사이트
# mil 미국의 군사 조직
# mobi 휴대 장치를 위한 사이트
# museum 미술관과 박물관
# name 개인 사용자
# net 네트워크를 관리하는 기관
# org 비영리 기관, 혹은 다른 gTLD에 해당하지 않는 단체. 역시 or가로 사용하기도 한다. ex)www.url.or.kr
# pro 회계사, 의사, 변호사 등의 전문가
# tel 전화 네트워크와 인터넷 사이의 연결을 관장하는 서비스
# travel 여행사, 항공사, 호텔 등
# wiki 위키
# xxx 포르노
# xyz
```
